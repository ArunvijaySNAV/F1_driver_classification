```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

print(np.__version__)
print(cv2.__version__)
```

```python
img = cv2.imread('/Users/snarunvijay/Documents/Machine_Learning/ImageClassification/model/dataset/lewis hamilton/12 memorable years for Lewis Hamilton with the Silver Arrows comes to end, but is he still hungry_ _ Motor-sport News - The Indian Express.jpg')
img.shape
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('/Users/snarunvijay/Documents/Machine_Learning/ImageClassification/model/opencv/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('/Users/snarunvijay/Documents/Machine_Learning/ImageClassification/model/opencv/haarcascades/haarcascade_eye.xml')
```

```python
print(hasattr(cv2, "CascadeClassifier"))
```

```python
faces = face_cascade.detectMultiScale(gray)
faces
```

```python
(x, y, w, h) = faces[0]
print(x, y, w, h)
```

```python
face_img = cv2.rectangle(img, (x, y), (x+h, y+w), (255, 0, 0), 2)
```

```python
plt.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for (x, y, w, h) in faces:
    face_img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```
